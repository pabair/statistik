## Normalverteilung

### Erzeugen von normalverteilten Zufallszahlen

Zuerst erzeugen wir 100000 normalverteilte Zufallszahlen mit Hilfe des `random`-packages und plotten ihre Verteilung in einem Histogram: 

```python
import random

numbers = []
for i in range(0, 100000):
    r = random.gauss(mu=0, sigma=1)
    numbers.append(r)
```

```python
import matplotlib.pyplot as plt
plt.hist(numbers, 50, rwidth=0.9)
plt.show()
```

Nun wollen wir mit Hilfe der 12er Regel selbst normalverteilte Zufallszahlen erstellen. Dafür addieren wir 12 Zufallszahlen aus einer Gleichverteilung.

```python
results=[]
for i in range(0, 100000):
    summe = 0
    for i in range(0, 12):
        summe += random.random()
    results.append(summe)
```

```python
plt.hist(results, 50, rwidth=0.9)
plt.show()
```

### Zentraler Grenzwertsatz

Im Folgenden schauen wir uns den zentralen Grenzwertsatz an, der besagt dass sich die Summe von beliebig verteilten Zufallsvariablen der Normalverteilung annähert.

Dazu definieren wir eine Funktion die `n`-Würfelwürfe aufaddiert:

```python
def throw_dice(n):
    results = []
    for i in range(0, 100000):
        sum_dice = 0
        for k in range(0, n):
            sum_dice += random.randint(0,6)
        results.append(sum_dice)
    return results
```

Und plotten die Verteilung für verschiedene `n`:

```python
n = 2
results = throw_dice(n)
plt.hist(results, 13, rwidth=0.9, density=True)
plt.show()
```

```python
n=5
results=throw_dice(n)
plt.hist(results, 13, rwidth=0.9, density=True)
plt.show()
```

```python
n=200
results=throw_dice(n)
plt.hist(results, 13, rwidth=0.9, density=True)
plt.show()
```

Wir können sehr, das mit größerem `n` die Verteilung mehr und mehr die Form einer Normalverteilung annimmt.
